# Exploratory Data Analysis

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("epl_final.csv")
print(df.shape)
print(df.dtypes)
print(df.head())
print(df.isnull().sum())
```

```python
df['FullTimeResult'].value_counts().plot(kind='bar')
plt.title('Match Outcome Distribution (H/D/A)')
plt.xlabel('Result')
plt.ylabel('Count')
plt.show()
```

```python
print(df['FullTimeResult'].value_counts(normalize=True))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
df['FullTimeHomeGoals'].hist(ax=axes[0], bins=10)  # Full time home goals
axes[0].set_title('Home Goals Distribution')
df['FullTimeAwayGoals'].hist(ax=axes[1], bins=10)  # Full time away goals
axes[1].set_title('Away Goals Distribution')
plt.show()
```

```python
df.groupby('Season')[['FullTimeHomeGoals', 'FullTimeAwayGoals']].mean().plot()
plt.title('Average Goals per Game by Season')
plt.ylabel('Goals')
plt.show()
```
